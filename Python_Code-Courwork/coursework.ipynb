```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Make the new graph named G
G = nx.Graph()

#Task 1: 6 stations with 1 line
G.add_nodes_from([
    ("A", {"Name": "Hyde Park Corner"}),
    ("B", {"Name": "Green Park"}),
    ("C", {"Name": "Piccadilly Circus"}),
    ("D", {"Name": "Leicester Square"}),
    ("E", {"Name": "Covent Garden"}),
    ("F", {"Name": "Holborn"})
])

# Add the edges of each node
G.add_edges_from([
    ("A", "B", {"Distance": 0.8}),
    ("B", "C", {"Distance": 0.8}),
    ("C", "D", {"Distance": 0.3}),
    ("D", "E", {"Distance": 0.3}),
    ("E", "F", {"Distance": 0.4})
])

# Node's position
position = {
    "A": (0, 0),
    "B": (1, 1),
    "C": (2.5, 1),
    "D": (3, 1),
    "E": (3.5, 1.5),
    "F": (4, 2)
}

# Station's name position
label_position = {
    "A": (0.5, 0),
    "B": (1, 1.15),
    "C": (2.35, 0.85),
    "D": (3, 1.15),
    "E": (3.5, 1.65),
    "F": (3.75, 2)
}

# Set the node and edge label
node_labels = nx.get_node_attributes(G, "Name")
edge_labels = {(u, v): f"{d['Distance']} km" for u, v, d in G.edges(data=True)}

plt.figure(figsize=(14, 7))

# Draw the graph 
nx.draw_networkx(G, pos=position, with_labels=False, node_size=200, node_color="blue", edge_color="blue")

# Print the station name to the graph
for p in position:
    x, y = label_position[p]
    plt.text(x, y, node_labels[p], fontsize=12, ha='center', va='center', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

# Print the distance edge label
nx.draw_networkx_edge_labels(G, pos=position, edge_labels=edge_labels, font_color='black')

# Add dummy line and legend
piccadilly_line = plt.Line2D([], [], color='blue', label='Piccadilly', linewidth=2)
plt.legend(handles=[piccadilly_line], loc='lower right', title="Key", fontsize=10, title_fontsize='12')

plt.show()
```

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Make the new graph named G
G = nx.Graph()

#Task 1: 6 stations with 1 line
G.add_nodes_from([
    ("A", {"Name": "Hyde Park Corner"}),
    ("B", {"Name": "Green Park"}),
    ("C", {"Name": "Piccadilly Circus"}),
    ("D", {"Name": "Leicester Square"}),
    ("E", {"Name": "Covent Garden"}),
    ("F", {"Name": "Holborn"}),
    ("G", {"Name": "Embackment"}),
    ("H", {"Name": "Charing Cross"}),
    ("I", {"Name": "Tottenham Court Road"}),
    ("J", {"Name": "Warren Street"}),
    ("K", {"Name": "Euston"}),
    ("L", {"Name": "Oxford Circus"}),
    ("M", {"Name": "Bond Street"}),
    ("N", {"Name": "Marble Arch"}),
    ("O", {"Name": "Lancaster Gate"}),
    ("P", {"Name": "Swiss Cottage"}),
    ("Q", {"Name": "St John's Wood"}),
    ("R", {"Name": "Baker Street"}),
    ("S", {"Name": "Westmenster"}),
    
])

# Node's position
position = {
    "A": (1, 0),
    "B": (1.7, 1),
    "C": (2.5, 1),
    "D": (3, 1),
    "E": (3.4, 1.75),
    "F": (3.8, 2.65),
    "G": (3, -1),
    "H": (3, 0),
    "I": (3, 2.75),
    "J": (3, 4),
    "K": (3.4, 4.7),
    "L": (2.25, 2.75),
    "M": (1.45, 2.75),
    "N": (0.8, 2.75),
    "O": (0.5, 1.75),
    "P": (1, 5.5),
    "Q": (1.45, 4.8),
    "R": (1.45, 4),
    "S": (2.5, -1),
}

# Station's name position
label_position = {
    "A": (1, -0.25),
    "B": (1.7, 1.25),
    "C": (2.5, 1.25),
    "D": (3.25, 1),
    "E": (3.4, 2),
    "F": (3.8, 2.85),
    "G": (2.8, -1),
    "H": (3, 0.25),
    "I": (3, 2.35),
    "J": (3, 4.5),
    "K": (3.5, 4.2),
    "L": (2.25, 3),
    "M": (1.45, 3),
    "N": (0.8, 3),
    "O": (0.5, 1.4),
    "P": (1, 5.8),
    "Q": (1.25, 4.75),
    "R": (1.25, 4),
    "S": (2.5, -1.3),
}

# Add the edges of each node
G.add_edges_from([
    ("A", "B", {"Distance": 0.8, "Line": "Piccadilly"}),
    ("B", "C", {"Distance": 0.8, "Line": "Piccadilly"}),
    ("C", "D", {"Distance": 0.3, "Line": "Piccadilly"}),
    ("D", "E", {"Distance": 0.3, "Line": "Piccadilly"}),
    ("E", "F", {"Distance": 0.4, "Line": "Piccadilly"}),
    ("G", "H", {"Distance": 0.8, "Line": "Goodge Street"}),
    ("H", "D", {"Distance": 0.5, "Line": "Goodge Street"}),
    ("D", "I", {"Distance": 1.3, "Line": "Goodge Street"}),
    ("I", "J", {"Distance": 0.6, "Line": "Goodge Street"}),
    ("J", "K", {"Distance": 0.5, "Line": "Goodge Street"}),
    ("O", "N", {"Distance": 0.3, "Line": "Hill Gate"}),
    ("N", "M", {"Distance": 0.8, "Line": "Hill Gate"}),
    ("M", "L", {"Distance": 0.6, "Line": "Hill Gate"}),
    ("L", "I", {"Distance": 0.5, "Line": "Hill Gate"}),
    ("I", "F", {"Distance": 0.9, "Line": "Hill Gate"}),
    ("P", "Q", {"Distance": 1.0, "Line": "Westminster"}),
    ("Q", "R", {"Distance": 0.3, "Line": "Westminster"}),
    ("R", "M", {"Distance": 0.7, "Line": "Westminster"}),
    ("M", "B", {"Distance": 1.0, "Line": "Westminster"}),
    ("B", "S", {"Distance": 1.3, "Line": "Westminster"}),
    
])

# Set the node and edge label
node_labels = nx.get_node_attributes(G, "Name")
edge_labels = {(u, v): f"{d['Distance']} km" for u, v, d in G.edges(data=True)}

plt.figure(figsize=(20, 10))

# Draw color
colors = {"Piccadilly": "blue", "Hill Gate": "purple", "Westminster": "red", "Goodge Street": "yellow"}
# a draw colour code
nx.draw_networkx(G, pos=position, with_labels=False, node_size=150)

# Print the station name to the graph
for u, v, d in G.edges(data=True):
    nx.draw_networkx_edges(G, pos=position, edgelist=[(u, v)], edge_color=colors[d['Line']], width=2)

# Drawing nodes and labels
nx.draw_networkx_nodes(G, pos=position, node_size=150,  node_color='white', edgecolors='black')
nx.draw_networkx_labels(G, pos=label_position, labels=node_labels, bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'), font_size=12)

#Print the distance edge label
nx.draw_networkx_edge_labels(G, pos=position, edge_labels=edge_labels, font_color='black',)

# Add dummy line and legend
dummy_lines = [plt.Line2D([], [], color=color, label=line) for line, color in colors.items()]
plt.legend(handles=dummy_lines, loc='lower right', title="Key", fontsize=10, title_fontsize='13')

# Task 3
# Get the distance of these nodes and store it in the list

total_length = 0
average_distance = 0
standard_deviation = 0

the_distance = []
for u, v, data in G.edges(data=True):
    the_distance.append(data["Distance"])

for x in the_distance:
    total_length += x
total_length = round(total_length, 3)
average_distance = round(total_length / len(the_distance), 3)
standard_deviation = round(np.std(the_distance), 3)

stats_text = (
    f"Total Length: {total_length} km\n"
    f"Average Distance: {average_distance} km\n"
    f"Standard Deviation: {standard_deviation}"
)
plt.text(0.02,0.1, stats_text, transform=plt.gca().transAxes, verticalalignment='top', fontsize=12)

plt.show()
```
